| Type de jointure | Description                                                              | Lignes retournées           |
| ---------------- | ------------------------------------------------------------------------ | --------------------------- |
| **INNER JOIN**   | Correspondances uniquement                                               | Lignes communes             |
| **LEFT JOIN**    | Tous les enregistrements de gauche + correspondances à droite            | Tous les éléments de gauche |
| **RIGHT JOIN**   | Tous les enregistrements de droite + correspondances à gauche            | Tous les éléments de droite |
| **FULL JOIN**    | Toutes les lignes des deux tables (avec `NULL` si pas de correspondance) | Toutes les lignes           |
| **SELF JOIN**    | Jointure d’une table avec elle-même                                      | Comparaisons internes       |


```python
join_query ="""
select MenuItems.Name,
MenuItems.type,
Menus.Cuisine,
MenuItems.Price from MenuItems
inner join Menus on MenuItems.ItemID =Menus.ItemID ;"""

cursor.execute(join_query)
result = cursor.fetchall();
cols = cursor.column_names;
print(cols)
for i in result:
    print(i)

    
```

```python
join_query =""" 
select MenuItems.Name,
MenuItems.Type,
Menus.Cuisine,
MenuItems.Price 
from MenuItems
left join Menus on MenuItems.ItemID = Menus.ItemID
where Cuisine is null  ;"""

cursor.execute(join_query)
result = cursor.fetchall()
col = cursor.column_names;
print(col)
for i in result:
    print(i)
```

```python
join_query=""" 
select Bookings.BookingID,
Bookings.TableNo,
Bookings.GuestFirstName,
Bookings.EmployeeID,
Orders.BillAmount
from Bookings
left join Orders on Bookings.BookingID = Orders.BookingID
  ;"""
cursor.execute(join_query) ;

result =cursor.fetchall();
cols =cursor.column_names
print(cols)
for i in result:
    print(i)
```

```python
sql_query ="""
select BookingID  AS ID,
UPPER(CONCAT(GuestFirstName, ' ',GuestLastName ))AS GuestFullName 
from Bookings
 ;"""
cursor.execute(sql_query) ;
result = cursor.fetchall();
cols = cursor.column_names;
print(cols)
for i in result:
    print(i)
```

```python
sql_query ="""select count(BookingID) as nombre_Booking,
SUM(BillAmount)as total_vente,
avg(BillAmount) as moyenne_total
from Orders ;"""
cursor.execute(sql_query) ;
result = cursor.fetchall();
cols = cursor.column_names ;
print(cols);
for i in result:
    print('nombre_Booking',i[0]);   
    print('total_vente',i[1]);    
    print('moyenne des vente',i[2]);
```

```python
sql_query = """select TableNo ,
count(TableNo) as Numero_Booking 
from Bookings 
GROUP BY TableNo 
ORDER BY Numero_Booking DESC;"""
cursor.execute(sql_query) ;
result = cursor.fetchall()
cols = cursor.column_names ;
print(cols);
for i in result:
    print(i)
```
